# scipy

```python
from scipy import stats
```

```python
rezultat = stats.linregress(suprafete, preturi)
```

```python
rezultat.slope, rezultat.intercept
```

```python
def f_optim(x):
    a = rezultat.slope
    b = rezultat.intercept
    return a * x + b
```

```python
f_optim(100)
```

```python
preturi_in_linie_1 =  1600 * suprafete
preturi_in_linie_2 =  1650 * suprafete + 2000

preturi_optime = 1655.1916376306622 * suprafete -1805.8536585365946

plt.scatter(suprafete, preturi);
# plt.plot(suprafete, preturi);
plt.plot(suprafete, preturi_in_linie_1, c="green");
plt.plot(suprafete, preturi_in_linie_2, c="red");

plt.plot(suprafete, preturi_optime, c="black");
```
