```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
apartamente = [(37, 65_000), (40, 68_000), (55, 85_000), (60, 90_000),  (80, 130_000), (120, 200_000)]
apartamente = np.array(apartamente)
apartamente
```

```python
suprafete = apartamente[:, 0]
preturi = apartamente[:, 1]
suprafete, preturi
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
x = suprafete
y = preturi
```

```python
x, x.shape
```

```python
reshaped_x = x.reshape((-1, 1))
reshaped_x, reshaped_x.shape
```

```python
model.fit(reshaped_x, y)
```

```python
valori_deprezis =  [
    [100], [114]
]

model.predict(valori_deprezis)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from scipy import stats
rezultat = stats.linregress(x, y)
rezultat
```

```python
preturi_optime = model.coef_ * suprafete + model.intercept_

plt.scatter(suprafete, preturi);

plt.plot(suprafete, preturi_optime, c="black");
```

```python


preturi_optime_scipy = rezultat.slope * suprafete + rezultat.intercept
plt.plot(suprafete, preturi_optime_scipy, c="red");


preturi_optime_sklearn = model.coef_ * suprafete + model.intercept_ 
plt.plot(suprafete, preturi_optime_sklearn, c="black");

plt.scatter(suprafete, preturi);

```
