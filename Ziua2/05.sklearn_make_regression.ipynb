```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = make_regression(n_samples=2, n_features=3)
len(data), data
```

```python
data = make_regression(n_samples=5, n_features=1)
len(data), data
```

```python
x, y = data
```

```python
plt.scatter(x, y);
```

```python
x, y = make_regression(n_samples=30, n_features=1)
plt.scatter(x, y);
```

```python
x, y = make_regression(n_samples=100, n_features=1)
plt.scatter(x, y);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
print(model)
```

```python
model.fit(x, y)
```

```python
y_prezis = model.coef_ * x + model.intercept_
plt.plot(x, y_prezis, c="red")
plt.scatter(x, y);
```
