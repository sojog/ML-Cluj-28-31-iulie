# TARGET VALUES

```python
y  = df["medv"]  # # cu 1 paranteza -> Serie (array unidimensional)
type(y )
```

# DATA -> features

```python
x = df.drop("medv", axis=1)
x
```

```python
x.shape, y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=8989)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
model = LinearRegression()
model
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_model = KNeighborsRegressor(n_neighbors=4)
knn_model
```

```python
knn_model.fit(x_train, y_train)
```

```python
knn_model.score(x_test, y_test)
```

```python
linear_y_predict = model.predict(x_test)
knn_y_predict = knn_model.predict(x_test)
```

```python
type(linear_y_predict), type(y_test), linear_y_predict.shape
```

```python
linear_y_predict = linear_y_predict.flatten()
knn_y_predict = knn_y_predict.flatten()
real_y = np.array(y_test).flatten()
```

```python
pd.DataFrame({"real": real_y,
              "linear":linear_y_predict, 
              "knn":knn_y_predict})
```
